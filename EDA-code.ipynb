```python
#calculating the average number of people per day in each station both getting on and getting off the trains
avg_ppl_perday=data[['05:00 - 06:00','06:00 - 07:00','07:00 - 08:00','08:00 - 09:00','09:00 - 10:00','10:00 - 11:00','11:00 - 12:00','12:00 - 13:00','13:00 - 14:00','14:00 - 15:00','15:00 - 16:00','16:00 - 17:00','17:00 - 18:00','18:00 - 19:00','19:00 - 20:00','20:00 - 21:00','21:00 - 22:00','22:00 - 23:00','23:00 - 24:00']].mean(axis=1)
print(avg_ppl_perday)
```

```python
#creating a new dataframe that consists of the dataset + extra column (column containing average number of people in the station per day)
df=data.assign(Average_no_people=lambda x: avg_ppl_perday)
```

```python
df.head()
```

Finding the train station that is the 'most crowded' and the 'least crowded' :

```python
#collecting all the station no.s and station names present in the dataset
station_no=[]
station_name=[]
station_no=data['Station No.'].unique()
station_name=data['Station Name'].unique()
print(station_no)
print(station_name)
```

```python
#calculating the average number of people in each station per day
all_avg=[]
l=len(data)
for i in station_no:
  station=df[df['Station No.']==i]
  sum=station['Average_no_people'].sum(axis=0)
  avg=sum/91
  all_avg.append(avg)
```

```python
print(all_avg)
```

```python
#building a dataframe consisting of all the
d={'Station no':station_no,'Station Name':station_name,'Average people':all_avg}
df1=pd.DataFrame(d)
df1
```

```python
#graph of station no. vs average number of people
g=plt.figure()
g.set_size_inches(10,6)
plt.scatter(df1['Station no'],df1['Average people'])
plt.xlabel('Station No.')
plt.ylabel('Average no. of people per day')
```

```python
#finding the details of the train station with the maximum number of average people per day (most crowded station)
maximum=df1['Average people'].max(axis=0)
most_crowded=df1[df1['Average people']==maximum]
most_crowded
```

```python
#finding the details of the train station with the minimum number of average people per day (least crowded station)
minimum=df1['Average people'].min(axis=0)
least_crowded=df1[df1['Average people']==minimum]
least_crowded
```

***Therefore, the most crowded train station in Daegu is 반월당2 (Banwoldang2) and the least crowded train station in Daegu is 학정 (Hakjeong)***
